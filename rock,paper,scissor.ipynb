```
#step one 
def playerChoice():
  playersChoice = 'swati'
  while not ( playersChoice == 'r' or playersChoice == 'p' or playersChoice == 's'):
    playersChoice = input("enter player's choice r p s - ")
    playersChoice = playersChoice.lower()
  return playersChoice
```

```
#step 2
import random 
def getComputerschoice():
  lst = ['r','p','s']
  return random.choice(lst)
```

```
#step3
def checkForwinner(player, computer):
  if player == 'r' and computer == 'r':
    return "draw"
  elif player == 'p'and computer == 'r':
    return "player"
  elif player == 's' and computer == 'r':
    return "computer"
  elif player == 'r' and computer == 'p':
    return "computer"
  elif player == 'p' and computer == 'p':
    return "draw"
  elif player == 's' and computer == 'p':
    return "player"
  elif player == 'r' and computer == 's':
    return "player"
  elif player == 'p' and computer == 's':
    return "computer"
  elif player == "s" and computer == 's':
    return "draw"
  else:
    return "draw"
```

```
#step4
def rpsDrivercode():
  rpsEndgame = "n"
  playersScore = 0 
  computerScore = 0 
  while rpsEndgame != 'y':

    ply = playerChoice()
    bot = getComputerschoice()
    print("Bot Entered - ", bot) 
    print(" ")

    win = checkForwinner( player = ply, computer= bot )
    print(win, " is the Winner")
    print(" ")

    if win == "player":
      playersScore = playersScore + 2
    elif win == "computer":
      computerScore  = computerScore + 2
    else:
      playersScore = playersScore + 1
      computerScore  = computerScore + 1
    print("Score Board ")
    print("Player - ", playersScore)
    print("Computer - ", computerScore)
    print(" ")
    rpsEndgame = input("Do you want to quit ? - (y/n) - ")

    
```
